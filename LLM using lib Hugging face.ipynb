```python
#Importando o Pipeline da LIb transformers
from transformers import pipeline
```

```python
# criando o pipeline de geração de textos com um modelo em português do Brasil
gerador = pipeline("text-generation", model="pierreguillou/gpt2-small-portuguese", device=0)
```

```python
# Prompt do texto de saida
prompt = " o primeiro presidente do Brasil foi"
```

```python
#Resposta apos a entrada do modelo
resposta = gerador(prompt, max_length=50, num_return_sequences=1)
```

```python
print(resposta[0]["generated_text"])
```
